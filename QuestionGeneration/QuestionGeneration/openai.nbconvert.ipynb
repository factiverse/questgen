```python
from nltk.translate.bleu_score import sentence_bleu
import nltk

# Sample reference and hypothesis sentences
reference = 'There exists a cat on the mat'
hypothesis = 'the cat is on the mat'

reference_tokenize = nltk.word_tokenize(reference)
hypothesis_tokenize = nltk.word_tokenize(hypothesis)
# Compute BLEU score

print(reference_tokenize, hypothesis_tokenize)

bleu_score = sentence_bleu([reference_tokenize], hypothesis_tokenize, weights=(1.0, 1.0, 1.0, 1.0))

# bleu_score = sentence_bleu([reference_tokenize], hypothesis_tokenize, weights=(0.25, 0.25, 0.25, 0.25))
# Compute ROUGE score
# rouge_score = rouge_n(hypothesis, reference[0], 2)

# Print scores
print(f'BLEU score: {bleu_score}')
# print(f'ROUGE score: {rouge_score}')
```
